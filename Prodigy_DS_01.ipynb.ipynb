```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read csv file 
df = pd.read_csv("C:\\Users\\Mega Store\\Documents\\jupyter\\task 1\\WorldPopulationByAge2020.csv" )
df.info()
df.head()
```

```python
# Distribution of the AgeGroup of population 
plt.figure(figsize=(12, 8))
plt.bar(df['AgeGroup'], df['population'], color="purple")
plt.xlabel('AgeGroup')
plt.ylabel('Population')
plt.title('Total Population vs Age Group ')
plt.xticks(rotation=90)
plt.show()
```
